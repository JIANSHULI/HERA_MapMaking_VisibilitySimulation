```python
import numpy as np
import healpy as hp
import astropy as ap
import matplotlib.pyplot as plt
from astropy.io import fits
NSIDE=32
m=np.arange(hp.nside2npix(NSIDE))
print m.shape, m[0:10], m[m.shape[0]-10:m.shape[0]]
hp.mollview(m, title="Test")
```

```python
hdul = fits.open('test.fits')
hdul.info()
hdr = hdul[0].header
data = hdul[0].data
hdr
print 'Data shape, min, max: ',data.shape, min(data), max(data)
testmap = fits.getdata('test.fits')
plt.hist(testmap,bins=100)
hp.mollview(testmap, title="Test Map")
```

```python
hp.cartview(testmap, title="Test Map", norm='log', min=10, max=1250, coord='C', latra=[-50,-10],xsize=360)
hp.graticule(dpar=10,dmer=10)
```
